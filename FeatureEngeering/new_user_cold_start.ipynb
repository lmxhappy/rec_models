```python
# 假设视频长度分别为 A_length, B_length, C_length
A_length = 100  # 示例视频 A 的长度
B_length = 150  # 示例视频 B 的长度
C_length = 200  # 示例视频 C 的长度

# 计算每个视频的倒数
A_score = 1 / A_length
B_score = 1 / B_length
C_score = 1 / C_length

# 找到倒数的最大值和最小值
min_score = min(A_score, B_score, C_score)
max_score = max(A_score, B_score, C_score)

# 对倒数进行归一化处理，使得分数在 [1, 2] 范围内
normalized_A_score = 1 + (1 - (A_score - min_score) / (max_score - min_score))
normalized_B_score = 1 + (1 - (B_score - min_score) / (max_score - min_score))
normalized_C_score = 1 + (1 - (C_score - min_score) / (max_score - min_score))

print("视频 A 的分数:", normalized_A_score)
print("视频 B 的分数:", normalized_B_score)
print("视频 C 的分数:", normalized_C_score)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def score_function(x, alpha):
    return 1+np.exp(-alpha * x)

# 视频长度范围
x_values = np.linspace(0, 30, 100)

# 不同 alpha 值对应的函数图像
alpha_values = [0.1, 0.2, 0.5, 1]

plt.figure(figsize=(10, 6))

for alpha in alpha_values:
    scores = score_function(x_values, alpha)
    plt.plot(x_values, scores, label=f'alpha={alpha}')

plt.title('Score Function for Video Length')
plt.xlabel('Video Length')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()
```
