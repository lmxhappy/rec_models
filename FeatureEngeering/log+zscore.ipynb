```python
import numpy as np
import matplotlib.pyplot as plt

# 设置参数
total_samples = 10000  # 总样本数
shape_parameter = 20.0  # 形状参数

# 生成长尾分布的随机数据
random_data = np.random.pareto(a=shape_parameter, size=total_samples)
```

```python
# 绘制直方图
plt.hist(random_data, bins=50, density=True, alpha=0.7, color='b')
plt.title('Long Tail Distribution')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()
```

```python
from scipy import stats

random_data_std = (random_data-random_data.mean())/random_data.std()
# 计算 z-score
random_data_zscore = stats.zscore(random_data)

```

```python
plt.hist(random_data_zscore, bins=50, density=True, alpha=0.7, color='b')
plt.title('Long Tail Distribution')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()
```

# 2 log+z_score

```python
rd_log = np.log(random_data)
```

```python
plt.hist(rd_log, bins=50, density=True, alpha=0.7, color='b')
plt.title('Long Tail Distribution')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()
```

```python
plt.hist(rd_log_zscore, bins=50, density=True, alpha=0.7, color='b')
plt.title('Long Tail Distribution')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()
```
